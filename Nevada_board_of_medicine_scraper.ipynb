```python
import requests
```

```python
# In this case, I used requests instead of urllib, as the later doesn't work with this html.
r = requests.get("http://medboard.nv.gov/Patients/Disciplinary/DisciplinaryActions/")
```

```python
from bs4 import BeautifulSoup
```

```python
document = BeautifulSoup(r.text, 'html.parser')
```

```python
#Take a look at the doctor's name list
h3_tag = document.find_all("h3")
tag_list = [tag.string for tag in h3_tag]
print tag_list
```

```python
# So it looks like all the content we need are in the div with class name: cat-questions
question_tags = document.find_all("div", attrs = {"class": "cat-questions"})
```

```python
# In the big div, each doctor has his/her own h3 and a div with class name "answer", which contants the action description from the board of medicine.
for tag in question_tags:
    div = tag.find_all("div")
    print div
```

```python
# See the level we need to parse in order to get just the content of the action.
for tag in question_tags:
    ul_tag = tag.find("ul")
    li_tag = ul_tag.find("li")
    ul_tag2 = li_tag.find("ul")
    li_tag2 = ul_tag2.find_all("li")
print li_tag2
```

```python
#Since the h3 and div"answer" is in the same level, we could parse them seperately and put the content into two lists.
#Name list
question_tags = document.find("div", attrs = {"class": "cat-questions"})
list2 = []
for h3_tag in question_tags.find_all("h3"):
    list2.append(h3_tag.string)
print len(list2)
```

```python
#Action list
list3 = []
for div in question_tags.find_all("div"):
    answer = div.find("ul")
    list3.append(answer)
print len(list3)
```

```python
#Combine two lists into a dictionary
doctor_dict = dict(zip(list2, list3))
```

```python
# Vola! Here is a dictionary with doctor's name as the key and the action description as the value.
doctor_dict
```

```python
#Now we need to export this dictionary into a dataframe to do more clearning work.
import pandas as pd
```

```python
pd.DataFrame(doctor_dict.items())
```

```python
df = pd.DataFrame(doctor_dict.items(), columns=['Name', 'Action'])
```

```python
# In my case, I just want to get a csv file and then do the cleaning part in excel.
df.to_csv("nevada_doctor.csv", index = False)
```
